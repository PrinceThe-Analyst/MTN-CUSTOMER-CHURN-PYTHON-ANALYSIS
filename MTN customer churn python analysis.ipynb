```python
import pandas as pd
```

```python
# which subscription plan has the highest churn rate?
```

```python
top_plan = churn_data.groupby("Subscription Plan") ["Total Revenue"].sum().sort_values(ascending = False).reset_index().head(5)
```

```python
top_plan
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.barplot(data = top_plan, x = "Subscription Plan", y = "Total Revenue")
plt.title("Subscription plan with high churn rate")
plt.xlabel("Subscription plan")
plt.ylabel("Total Revenue")
plt.ticklabel_format(style = 'plain', axis = 'y')
```

```python
# which customer age group churned the most?
```

```python
top_age = churn_data.groupby("Age") ["Total Revenue"].sum().sort_values(ascending = False).reset_index().head(5)
```

```python
top_age
```

```python
sns.barplot(data = top_age, x = "Age", y = "Total Revenue")
plt.title("Age with highest churn")
plt.xlabel("Age range")
plt.ylabel("Total Revenue")
plt.ticklabel_format(style = 'plain', axis = 'y')
```

```python
# which states has the highest record of churned customers?
```

```python
top_states = churn_data.groupby("State") ["Total Revenue"].sum().sort_values(ascending = False).reset_index().head(5)
```

```python
top_states
```

```python
sns.barplot(data = top_states, x = "Total Revenue", y = "State")
plt.title("States with highest churn rate")
plt.xlabel("Total Revenue")
plt.ylabel("State")
plt.ticklabel_format(style = 'plain', axis = 'x')
```

```python
# what are the most common reasons for customer churn?
```

```python
churn_reason = churn_data.groupby("Reasons for Churn") ["Total Revenue"].sum().sort_values(ascending = False).reset_index().head(5)
```
